```python
def cn_ci(dir_path):
    for rdf in ci_list:
        jieba.add_word(rdf[0])
    all_text = u""
    for file_name in os.listdir(dir_path):
        if file_name.find(".txt") != -1:
            file_path = "/".join([dir_path, file_name])
            with open(file_path, "r") as f:
                all_text += f.read()

    terms = jieba.cut(all_text)

    return [ci for ci in ','.join(terms).split(',') if ci not in [u'', u" "]]
# this function is use for cut word in the word list
```

```python
# this function below is use to statist frequency of every word in the list

def cn_train(features):
    model = collections.defaultdict(lambda: 1)
    for f in features:
        model[f] += 1
    return model
CNWORDS = cn_train(cn_ci("cn_texts"))
```

```python
cn_words = pd.DataFrame(columns = ("word", "frequency"), index = None)
for i,j in enumerate(list(CNWORDS.items())):
    cn_words.loc[i] = [j[0],j[1]]
    if i%2000 == 0:
        print("200 words have done.")
```

```python
cn_words = cn_words.sort("frequency", ascending=False)
cn_words = cn_words.reset_index()
```

```python
cn_words = cn_words[["word","frequency"]]
```

```python
cn_words.loc[cn_words["frequency"] > 400,'frequency'] = 400
```

```python
# import all word

def cn_hanzi():
    with open("hanzi.txt","r") as f:
        hanzi = f.read()
        return hanzi
cn_hanzi_str = cn_hanzi()
```

```python
# construct possible world list, caculate distance 
def cn_edits1(ci):
    splits = [(ci[:i], ci[i:]) for i in range(len(ci)+3)]
    deletes = [a + b[1:] for a, b in splits if b if a + b[1:] in CNWORDS] 
    transposes = [a + b[1] + b[0] + b[2:] for a, b in splits if len(b)>1 if a + b[1] + b[0] + b[2:] in CNWORDS] 
    replaces = [a + c + b[1:] for a, b in splits for c in cn_hanzi_str if b if a + c + b[1:] in CNWORDS]
    inserts = [a + c + b     for a, b in splits for c in cn_hanzi_str if a + c + b in CNWORDS]
    return set(deletes + transposes + replaces + inserts)
```

```python
def cn_known_edits2(ci):
    return set(e2 for e1 in cn_edits1(ci) for e2 in cn_edits1(e1) if e2 in CNWORDS)
```

```python
def cn_correct(ci):
    # candidatas list
    candidates = cn_edits1(ci) or cn_known_edits2(ci)
    # find the max frequency
    return max(candidates, key=CNWORDS.get)
```

```python
s = "再 生育 时会 有 25 % 的 风险 几率 生育 耿儿".split(" ")
s
```
